### Milestone - Data Extraction, Cleaning, and Loading

```python
import pandas as pd
import zipfile
import numpy as np
```

```python
with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
moma_artworks = pd.read_csv("Artworks.csv")
moma_artists = pd.read_csv("Artists.csv")
gender = moma_artists[["Gender", 'Name', 'Nationality']]
moma = moma_artworks.merge(gender, on = ['Name'], how = 'left')
moma
```

```python
no_date = moma[(moma['Acquisition Date'].isnull() == True)].index
moma.drop(no_date, inplace = True)
#dropped dates that didn't have an acquisition date
moma
```

```python
# create a new dataframe for the purpose of analyzing dates
moma_date = moma

# remove c., before, and after from the dates
# take the first four characters as the date
moma_date["Date"] = moma_date.Date.str.strip("c. ")
moma_date["Date"] = moma_date.Date.str.strip("before")
moma_date["Date"] = moma_date.Date.str.strip("after ")
moma_date['Date'] = moma_date.Date.astype(str).str[:4]

# remove all of the entries that have an unknown date
n_d = moma_date[(moma_date["Date"] == "nan") | (moma_date["Date"] == "n.d") | (moma_date["Date"] == "Unkn") | (moma_date["Date"] == "Unknown")].index
moma_date.drop(n_d, inplace = True)

# convert the strings back to integers
moma_date['Date'] = pd.to_numeric(moma_date['Date'], errors="coerce")
moma_date
```
